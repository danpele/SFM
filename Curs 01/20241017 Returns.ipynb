```python
import yfinance as yf

import numpy as np
import pandas as pd
#define the ticker symbol
tickerSymbol = 'MSFT'

#get data on this ticker
tickerData = yf.Ticker(tickerSymbol)

#get the historical prices for this ticker
df = tickerData.history(period='1d', start='2010-1-1', end='2024-1-25')

#see your data
df
```



Simple return: $R_t=\frac{P_t - P_{t-1}}{P_{t-1}}$
$\\$
Log-return: $r_t = \log\left(\frac{P_t}{P_{t-1}}\right)=\log{P_t}-\log{P_{t-1}}$

```python
close = df["Close"]

simple_return = close.pct_change()

log_return = np.log(1+simple_return)
```

```python
# Import the necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# This line ensures plots appear in Jupyter notebooks
%matplotlib inline

# Set the default figure size for the plots
plt.rcParams['figure.figsize'] = (15, 9)

# Plot the simple return and log return on the same plot
simple_return.plot(label="Simple Return", grid=True)
log_return.plot(label="Log Return", grid=True)

# Add a legend to differentiate between the two plots
plt.legend()

# Set the figure and axis backgrounds to transparent
plt.gcf().patch.set_alpha(0)  # Set figure background to transparent
plt.gca().patch.set_alpha(0)  # Set axis background to transparent

# Save the plot with a transparent background
plt.savefig('returns_plot.png', transparent=True)

# Show the plot
plt.show()
```

```python
plt.plot(simple_return,log_return, 'o',c='blue')
plt.ylabel( 'Log-return' )
plt.xlabel( 'Simple return' )
#plt.legend(loc='upper center',bbox_to_anchor=(0.5, -0.15), fancybox=True, shadow=True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot both log_return and simple_return on the same chart using histplot
sns.histplot(log_return, kde=False, stat='density', label='Log-returns', color='blue')
sns.histplot(simple_return, kde=False, stat='density', label='Simple-returns', color='orange')

# Add legend to identify each plot
plt.legend()
plt.xlabel( 'Returns' )
# Save and show the plot with a transparent background
plt.savefig('transparent_chart.png', transparent=True)
plt.show()

```

```python
# prompt: compute skewness and kurtosis of log_return

# Calculate skewness and kurtosis of log_return
skewness = log_return.skew()
kurtosis = log_return.kurt()

print(f"Skewness of log_return: {skewness}")
print(f"Kurtosis of log_return: {kurtosis}")
```
